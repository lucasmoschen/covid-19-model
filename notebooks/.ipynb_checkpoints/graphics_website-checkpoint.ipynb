# Graphics for the website 

Generate the graphics for the scenarios in the website. 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
scenario1 = pd.read_csv('../data/variables/scenario1_web.txt', index_col=0).transpose()
scenario2 = pd.read_csv('../data/variables/scenario2_web.txt', index_col=0).transpose()
scenario3 = pd.read_csv('../data/variables/scenario3_web.txt', index_col=0).transpose()
scenario4 = pd.read_csv('../data/variables/scenario4_web.txt', index_col=0).transpose()
```

```python
death1 = scenario1['D']
death2 = scenario2['D']
death3 = scenario3['D']
death4 = scenario4['D']

deaths = pd.DataFrame({'Scenario 1': death1, 
                       'Scenario 2': death2, 
                       'Scenario 3': death3, 
                       'Scenario 4': death4,
                       't': death4.index})
```

```python
cumulative1 = 1 - (scenario1['Sf'] + scenario1['Sr'])
cumulative2 = 1 - (scenario2['Sf'] + scenario2['Sr'])
cumulative3 = 1 - (scenario3['Sf'] + scenario3['Sr'])
cumulative4 = 1 - (scenario4['Sf'] + scenario4['Sr'])


cumulatives = pd.DataFrame({r'Scenario 1: $\rho=0.01$': cumulative1, 
                            r'Scenario 2: $\rho=0.03$': cumulative2, 
                            r'Scenario 3: $\rho=0.05$': cumulative3, 
                            r'Scenario 4: $\rho=0.1$': cumulative4,
                            't': cumulative4.index})
```

```python
fig, ax = plt.subplots(figsize = (12,8))
cumulatives.plot(logy = True, 
                 ax = ax, 
                 title = 'Cumulative infections during the pandemic'
                )
plt.savefig('../images/infections_website.png', bbox_inches='tight')
plt.show()

fig, ax = plt.subplots(figsize = (12,8))
deaths.plot(logy = True, 
                 ax = ax, 
                 title = 'Cumulative deaths during the pandemic', 
                )
plt.savefig('../images/deaths_website.png', bbox_inches='tight')
plt.show()
```
