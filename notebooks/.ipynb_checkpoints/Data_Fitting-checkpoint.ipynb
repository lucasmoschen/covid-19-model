## Hipotesis for the initial conditions

1. Let $x$ be the number of confirmed cases in the initial period and $H$ total number of inhabitants of the place: So $Q_0 = \frac{x}{H}$

2. The effective number of infected cases is:  $ I_{unreported, 0} = I_{u,0} = F*Q_0$, where $F$ is a factor number. 

3. $I_{u,0} = I_{f,0} + I_{r,0} + E_{f,0} + E_{r,0} + A_{f,0} + A_{r,0}$

4. $E_0 = k_{E}\cdot I_{u,0}$, $I_0 = k_{I}\cdot I_{u,0}$ and $A_0 = k_{A}\cdot I_{u,0}$, where $k_{E} + k_I + k_A = 1$
So $k_E = \frac{\tau}{\tau + \sigma + \gamma_1}, k_I = \frac{\sigma}{\tau + \sigma + \gamma_1}, k_A = \frac{\gamma_1}{\tau + \sigma + \gamma_1}$

5. $S_0 = (H - I_0 - E_0 - A_0 - Q_0 - R_0 - D_0)/H$

## Reading Rio de Janeiro Data: 16/03/2020 - 17/06/2020

```python
# Reading data
rio_detail_data = pd.read_csv("../data/oficial_rio_data20200812.csv", encoding = 'latin', sep = ";")
# Drop nan values in data notificatios column
rio_not_null = rio_detail_data.dropna(subset = ['dt_notific', 'dt_inicio_sintomas'])
# Creating pattern in data
notifications_dt = pd.to_datetime(rio_not_null['dt_notific'], format = '%d/%m/%Y')
symptoms_dt = pd.to_datetime(rio_not_null['dt_inicio_sintomas'], format = '%d/%m/%Y')
# Defining initial and final points
first_day = pd.to_datetime('2020-03-16')
last_day_nf = max(notifications_dt)
last_day_sy = max(symptoms_dt)

before_cases_nf = len(notifications_dt[notifications_dt < first_day])
before_cases_sy = len(symptoms_dt[symptoms_dt < first_day])
acumulated_test = {}
symptoms = {}
day = 0
while first_day + pd.Timedelta(days = day) <= last_day_nf:
    cases = len(notifications_dt[notifications_dt <= first_day + pd.Timedelta(days = day)])
    acumulated_test[first_day + pd.Timedelta(days = day)] = cases 
    day += 1

day = 0
while first_day + pd.Timedelta(days = day) <= last_day_sy:
    cases = len(symptoms_dt[symptoms_dt <= first_day + pd.Timedelta(days = day)])
    symptoms[first_day + pd.Timedelta(days = day)] = cases 
    day += 1
acumulated_test = pd.Series(acumulated_test)
symptoms = pd.Series(symptoms)
#rio_not_null.head()
```

## Reading BrasilIo Rio Data: 06/03/2020 - 16/06/2020

```python
rio_timeline_data = pd.read_csv("../data/brasilio_rio_data.csv")
#rio_timeline_data.head()
```

## Reading Brazil Data: 28/03/2020 - 16/05/2020

```python
brazil_data = pd.read_csv("../data/brazil_data.csv")
confirmed = brazil_data[brazil_data["municipio"] == "Rio de Janeiro"][["data","casosAcumulado"]]
confirmed = confirmed.set_index('data')
confirmed.index = pd.to_datetime(confirmed.index)
death = brazil_data[brazil_data["municipio"] == "Rio de Janeiro"][["obitosAcumulado","data"]]
death = death.set_index('data')
death.index = pd.to_datetime(death.index)
#brazil_data[brazil_data["municipio"] == "Rio de Janeiro"].head()
```

We are doing moving average of 7 days around the mean. 

```python
acumulated_test_7 = acumulated_test.rolling(window = 7, center = True).mean()
symptoms_7 = symptoms.rolling(window = 7, center = True).mean()

deaths = rio_timeline_data[['date','deaths']].set_index('date').iloc[::-1]
deaths.index = pd.to_datetime(deaths.index)

death_7 = deaths.rolling(window = 7, center = True).mean()
```

```python
symptoms_dt = symptoms.diff().iloc[1:]
#symptoms_dt = symptoms_dt.rolling(window=7, center = True).mean()
seasonal_filter = seasonal_decompose(symptoms_dt, model = 'multiplicative', period = 7)

data = symptoms_dt/(seasonal_filter.seasonal*seasonal_filter.resid)
```

## Experiments (Section devoted to make experiments)

```python
par_estimated = {"tau": 0.3125, "sigma": 1/2, "delta": 1.0, "gamma1": 1/9.5, "gamma2": 1/18,  
                 "mu": 0.1267/14, "rho": 2e-5, "r": 0.1, 
                 "p": 0}

# (name: [lb, ub, p0])
BI = {"F": [0.0, 10.0, 8.0], "p": None, "tau": None, "delta": None, "sigma": None, 
      "alpha": [0.7, 0.95, 0.9], 
      "beta": [0.05]*4 + [0.3]*4 + [0.1]*4,
      "r": None, "rho": None, "gamma1": None, "gamma2": None, "mu": None}

initial_day = '2020-03-21'
final_day = '2020-07-15'

outer_parameter = {'knots': np.linspace(0, len(data[initial_day:final_day]), 10)[1:-1], 
                   'bspline_order': 3,
                   'number_spline_coefficients': 4}
```

```python
%%time 
# len(knots) = number_spline_coefficients + bspline_order + 1
# 2**bspline_order <= knots

curve = Experiment('curve_fit', par_estimated, BI, data, 
                    Q0 = symptoms_7[initial_day], R0 = 0, D0 = death_7['deaths'][initial_day], 
                    outer_parameters = outer_parameter,
                    population = 6e6, initial_day = initial_day, final_day = final_day)
```

## Bootstrap

```python
data_param = {}
for i in range(0,4):
    with open('../experiments/experiments_bootstrap_22_09_2020-' + str(i) + '.json', 'r') as f: 
        datadict = json.load(f)
        datadict = {int(key) + i*75: datadict[key] for key in datadict.keys()}
        data_param.update(datadict)

F_values = np.array([data_param[key][0][0] for key in data_param.keys()])
beta1 = np.array([data_param[key][0][2] for key in data_param.keys()])
beta2 = np.array([data_param[key][0][3] for key in data_param.keys()])
beta3 = np.array([data_param[key][0][4] for key in data_param.keys()])
beta4 = np.array([data_param[key][0][5] for key in data_param.keys()])
alpha = np.array([data_param[key][0][1] for key in data_param.keys()])


fig, ax = plt.subplots(2,3, figsize = (18, 10))
sns.distplot(F_values, ax = ax[0][0], kde=False)
ax[0][0].set_title('F')
sns.distplot(beta1, ax = ax[0][1], kde=False)
ax[0][1].set_title('beta 1')
sns.distplot(beta2, ax = ax[0][2], kde=False)
ax[0][2].set_title('beta 2')
sns.distplot(beta3, ax = ax[1][0], kde=False)
ax[1][0].set_title('beta 3')
sns.distplot(beta4, ax = ax[1][1], kde=False)
ax[1][1].set_title('beta 4')
sns.distplot(alpha, ax = ax[1][2], kde=False)
ax[1][2].set_title('alpha')
plt.show()
```

```python
sns.pairplot(parameters_df)
plt.show()
```

```python
n = len(beta1)

par_estimated = {"tau": 0.3125, "sigma": 1/2, "delta": 1.0, "gamma1": 1/9.5, "gamma2": 1/18,  
                 "mu": 0.1267/14, "rho": 2e-4, "r": 0.1, 
                 "p": 0}

initial_day = '2020-03-21'
final_day = '2020-07-15'

outer_parameter = {'knots': np.linspace(0, len(data[initial_day:final_day]), 10)[1:-1], 
                   'bspline_order': 3,
                   'number_spline_coefficients': 4}

timeline = np.array(range(0, len(data[initial_day:final_day])))

curve = FitCovidCurve(par_estimated, data[initial_day:final_day]/(6e6), timeline, 6e6, 
                      Q0 = symptoms_7[initial_day], R0 = 0, D0 = death_7['deaths'][initial_day], 
                      outer_parameters = outer_parameter)

rtmatrix = np.empty(shape = (n, len(timeline)))
curvematrix = np.empty(shape = (n, len(timeline)))

for r in range(n): 
    parameters = dict(par_estimated)
    parameters['F'] = F_values[r]
    parameters['beta'] = [beta1[r], beta2[r], beta3[r], beta4[r]]
    parameters['alpha'] = alpha[r]
    curvematrix[r,:] = curve.integrate(timeline, parameters)
    rtlist = curve.repro_number()
    rtmatrix[r, :] = rtlist[1,:]
```

```python
fig, ax = plt.subplots(1,2, figsize = (24,8))
ax[0].fill_between(x = timeline, y1 = np.quantile(curvematrix, q = 0.975, axis = 0), 
                              y2 = np.quantile(curvematrix, q = 0.025, axis = 0), 
                              color = 'grey', alpha = 0.6)

ax[1].fill_between(x = timeline, y1 = np.quantile(rtmatrix, q = 0.025, axis = 0), 
                              y2 = np.quantile(rtmatrix, q = 0.975, axis = 0), 
                              color = 'grey', alpha = 0.6)

ax[1].plot(np.median(rtmatrix, axis = 0), color = 'red', label = 'Median')
ax[1].hlines(y = 1.0, xmin = timeline[0], xmax = timeline[-1], linestyle = '--', color = 'black')
ax[0].plot(np.median(curvematrix, axis = 0), color = 'red', label = 'Median')

ax[0].set_title('Curve')
ax[1].set_title('Rt')

ax[0].set_xlabel('Timeline')
ax[1].set_xlabel('Timeline')

ax[0].scatter(timeline, data[initial_day:final_day]/(6e6), color = 'darkblue', label = 'Real Data')

ax[0].legend(loc = 'upper left')
ax[1].legend()

fig.savefig('Curve_fitting.png', bbox_inches='tight')
plt.show()
```

## Why some have big error?

With this example we see how the p0 can influence the result!

```python
data = {}
for i in range(0,4):
    with open('../experiments/experiments_bootstrap_test' + str(i) + 'inside' + '.json', 'r') as f: 
        datadict = json.load(f)
        datadict = {int(key) + i*75*18: datadict[key] for key in datadict.keys()}
        data.update(datadict)
```

```python
errors = sorted(data.items(), key = lambda x: x[1][1], reverse=True)
```

```python
IB = errors[1][1][0][:-6]
p0 = errors[1][1][0][-6:]
error = errors[1][1][1]
IB_order = sorted(IB)
```

```python
p0[3]
```

```python
p0[3] = 0.06595597077201369
```

```python
par_estimated = {"tau": 0.3125, "sigma": 1/2, "delta": 1.0, "gamma1": 1/9.5, "gamma2": 1/18,  
                 "mu": 0.1267/14, "rho": 2e-4, "r": 0.1, 
                 "p": 0}

BI = {"F": [4.0, 10.0, 8.0], "p": None, "tau": None, "delta": None, "sigma": None, 
      "alpha": [0.7, 0.95, 0.9], 
      "beta": [0.05]*4 + [0.2]*4 + [0.1]*4,
      "r": None, "rho": None, "gamma1": None, "gamma2": None, "mu": None}

initial_day = '2020-03-21'
final_day = '2020-07-15'

data = symptoms_dt[initial_day:final_day]/(6e6)

def get_data_from_index(arr: np.array, indexer: np.array):
    X = np.array([arr[index] for index in indexer])
    return X

XB_order = get_data_from_index(data, IB_order)

curve = FitCovidCurve(par_estimated, XB_order, IB_order, 6e6, 
                      Q0 = symptoms_7[initial_day], R0 = 0, D0 = death_7['deaths'][initial_day], 
                      outer_parameters = {'bspline_order': 3, 'knots': np.linspace(0, len(data) - 1, 10)[1:-1], 'number_spline_coefficients': 4})

lb = [4.0, 0.7, 0.02, 0.02, 0.02, 0.02]
ub = [10.0, 0.95, 0.4, 0.4, 0.4, 0.4]
            
param, parameters, _ , T, _ = curve.fit('curve_fit', p0, lb, ub, sigma = None, absolute_sigma = None, rt_calc = False)
```

```python
plt.plot(data.index, T)
plt.plot(data)
```

```python
IO = list(range(len(data))) 
for i in set(IB):
    IO.remove(i)
```

```python
Y = curve.integrate(list(range(len(data))), parameters)
YO = get_data_from_index(Y, IO)
XO = get_data_from_index(data, IO)
error = (100*np.linalg.norm(XO - YO))/(100*np.linalg.norm(YO))
error
```
